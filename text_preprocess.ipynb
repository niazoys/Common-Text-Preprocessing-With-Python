```python
import numpy as np 
import pandas as pd 
import os
import nltk 
import string
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk import word_tokenize
from nltk.stem.wordnet import WordNetLemmatizer
from sklearn.utils import shuffle
from sklearn.neural_network import MLPClassifier

from sklearn.utils import shuffle
```

```python
#Load CSV File 
df = pd.read_csv('air.csv')

df.head()
```

```python
#Drop Null rows
df=df[pd.notnull(df['content'])]


#Make Lower Case 
df.content = df['content'].str.lower()

#Pull out numeric Values from review Text and Summary
df['content'] = df['content'].str.replace('\d+', '')

df.head()
```

```python
#Remove Punctuation
#Token Generate 
def tokenize(text):
    tokenized = word_tokenize(text)
    no_punc = []
    for review in tokenized:
        line = "".join(char for char in review if char not in string.punctuation)
        no_punc.append(line)
    tokens = lemmatize(no_punc)
    return tokens

#Lemmatization
#Stemming
def lemmatize(tokens):
    lmtzr = WordNetLemmatizer()
    stm = PorterStemmer()
    lemma = [lmtzr.lemmatize(t) for t in tokens]
    return lemma
```

```python

#Tokenize ReviewText and Summary
df.content = df.content.apply(lambda x: tokenize(x))

#Remove Stop-words
stops = stopwords.words('english')
df.content=df.content.apply(lambda x: [item for item in x if item not in stops])


df.head(5)

```
